## Introduction

For our final project, we have decided to leverage some of the amazing data resources online for the Bechdel test. This test was invented comic Alison Bechdel in 1985 in her [comic strip](https://upload.wikimedia.org/wikipedia/en/b/bf/Dykes_to_Watch_Out_For_%28Bechdel_test_origin%29.jpg) and is considered an indicator of gender equality in fiction, in our case specifically films. The three rules that must be passed are as follows:

1. It has to have at least two named women in it
2. Who talk to each other
3. About something besides a man

After investigating, we discovered there are many datasets that utilize resources like bechdeltest.com so we decided we wanted to create a more unique approach. We wanted to learn how LGBT films (as per tagged on IMDb) compare to non-LGBT films in terms of their Bechdel score. We also wanted to investigate how other variables like box office earnings and the gender of directors impact the Bechdel score result. Our three research questions are below:

1. Are box office earnings related to Bechdel test scores?
2. Do women more direct films that score higher on the Bechdel test than men?
3. Do LGBT films have a higher Bechdel test score than non-LGBT films?

The stages of our intestigation are 1) use the base Bechdel API to gather the scores (0-3), 2) scrape IMDb pages for LGBT tags to deterime if the films are "gay" or not, 3) call an open-source IMDb API to gather more film information like box office earrings, 4) run the director names through the [gender-guesser](https://pypi.org/project/gender-guesser/) Python machine learning (ML) package for gender recoginiton, 5) data cleaning, 6) feature engineering, and finally 7) visualizations to help answer our questions.

### Gender recognition ML
In order to make sense of the Directors and Actors data, we run a gender recognition ML tool on the dataset.  

The ML tool takes the first name (fails on full name, so we take the name before the first white space) as an argument and returns wether it is (more likely) male or female, androgynous (can't decide), or unknown in their dictionary.  

The director's gender is important for our research question.  
The Actor's gender is practically reflected in the Bechdel Test, we rather use it to control for gay-male romance films that naturally fail on the Bechdel-test by design.  

We won't care about N/A here, if the ML fails to recognise the name, then unknown is anyways going to be treated the same way.  

```python
d = gen.Detector()

for k , v in clean_dict.items():
    v['DirectorGender'] = [ d.get_gender( i.split()[0]) for i in v['Director'] ]
    v['ActorGender'] = [ d.get_gender( i.split()[0]) for i in v['Actors'] ]
```

Summarise the results into single variables.  

Director gender: if at least one of the directors is (more likely) female, it takes the value 'female'. Otherwise, if at least one is (more likely) male, then 'male', if everyone is unknown or ambigous angrogyn, then 'N/A'.

Actors' gender: if ALL (more likely) male, then 'male'. Otherwise if at least one is identified as (more) female, then 'female', for the rest 'N/A'.

```python
for k , v in clean_dict.items():
    dir_male = sum( [ ( i in ['male' , 'mostly_male']) for i in v['DirectorGender'] ] )
    dir_female = sum( [ ( i in ['female' , 'mostly_female']) for i in v['DirectorGender'] ] )
    
    if dir_female > 0:
        v['DirectorGender'] = 'female'
    elif dir_male > 0:
        v['DirectorGender'] = 'male'
    else: 
        v['DirectorGender'] = 'N/A'
        
    actor_male = sum( [ ( j in ['male' , 'mostly_male']) for j in v['ActorGender'] ] )
    actor_female = sum( [ ( j in ['female' , 'mostly_female']) for j in v['ActorGender'] ] )
    
    if actor_male == len( v['ActorGender'] ):
        v['ActorGender'] = 'male'
    elif actor_female > 0:
        v['ActorGender'] = 'female'
    else: 
        v['ActorGender'] = 'N/A'
```

We've confirmed our primary variables (bechdel_rating and IsItGay) have no null values which is great. There are some null values in the other variables though so we did the following:

1. **Replace with Sample Mean:** For any float/int variables, we used the sample mean and created a binary flag variable as an indicator (1 meaning average used and 0 meaning otherwise). This impacted the following: runtime, imdbVotes, and BoxOffice. For box office, we decided to not replace simply with the sample mean but instead with the decade specific sample mean to avoid placing modern box office values on historical films.
2. **Drop Values:** For any categorical or dateTime variables with <.5% of total rows, we decided to drop these values. This impacted the following: Released, Country, imdbRating, and CoProduction.
3. **Fill with Value:** For any binary values, we decided to fill with an Unknown value ('Unknown') to avoid lost observations. This impacted the following: DirectorGender and ActorGender.

```python
#Creating flag variable and replacing with Sample Mean
#sample_mean_cols = ['Runtime', 'imdbVotes', 'BoxOffice']
for cols in clean_df[['Runtime', 'imdbVotes', 'BoxOffice']]:
    clean_df[cols + '_NaN' ] = clean_df[cols].isnull()*1
    mean_value = clean_df[cols].mean(skipna = True)
    clean_df[cols].fillna(value=mean_value, inplace=True)
    print(cols)
```

```python
#Checking to make sure it got rid of the NaNs
clean_df.isnull().sum(axis = 0)
```

```python
#Moving onto Dropping Values
for cols in clean_df[['Released', 'Country', 'imdbRating', 'CoProduction']]:
    clean_df = clean_df[pd.notnull(clean_df[cols])]
    print(cols)
```

```python
#Again checking to make sure it got rid of the NaNs
clean_df.isnull().sum(axis = 0)
```

```python
#Now finally filling with value
for cols in clean_df[['DirectorGender', 'ActorGender']]: 
    clean_df[cols] = clean_df[cols].fillna('Unknown')
    print(cols)
```

```python
q2plot1 = sns.countplot(data = final_df, x='bechdel_rating', hue='DirectorGender')
q2plot1.set_xlabel("Bechdel Rating")
q2plot1.set_ylabel("Count of Films")
```

As we can see from this simple barchart, it appears females are more likely to direct films that pass the Bechdel test. However, there are so many male directors that it is not very clear exactly *how* much more likely films by females are to pass the Bechdel test. Let's create a heatmap for even easier visualizations.

```python
#Creating a pivot table for data to plot for heat map
final_df_q2 = pd.pivot_table(final_df, values=['imdbid'], 
                     index=['bechdel_rating'], 
                     columns=['DirectorGender'], aggfunc='count')

genders = ['Unknown', 'female', 'male']

for i in genders:
    final_df_q2[('imdbid', i)] = (final_df_q2[('imdbid', i)] /final_df_q2[('imdbid', i)].sum()) * 100
    final_df_q2.rename(columns={('imdbid', i): i}, inplace=True)
    

print(final_df_q2)
```

```python
q2plot2 = sns.heatmap(final_df_q2, cmap='PiYG')
q2plot2.set_xticklabels(['Unknown','Female','Male'])
q2plot2.set_xlabel("Director Gender")
q2plot2.set_ylabel("Bechdel Rating")
```

As we can see from this heatmap, female directors are *much* more likely to direct films that pass the Bechdel test. In fact, 79% of films directed by females pass the Bechdel test compared to 53% of films directed by males. 

#### Question 3. Do LGBT films have a higher Bechdel test score than non-LGBT films?
Let's move onto our third and final question based on the LGBT tags we so carefully scraped from the IMDb pages. Do films with the LGBT tags score higher on the Bechdel test than non-LGBT films? We used our "IsItGay" variable to investigate. Since these are 2 binary variables, we can use a similiar logit model as in our first question.

```python
sns.lmplot(x = "IsItGay", y = "bechdel_binary",
                        data = final_df,
                        y_jitter=.02, 
                        logistic=True,
                        col="DirectorGender", 
                        hue="DirectorGender",
                        truncate=False)
```

Running a similar logit model as we did for box office earnings, we can see if a film is tagged as LGBT on IMDb, it is more likely to pass the Bechdel test. This is true across both genders of directors including the "unknown" gender category. However, let's dive into really how much these LGBT tags impact the Bechdel score. We are going to try using a parallel categories visualization to understand the connections between the "IsItGay" and "bechdel_binary" variable and also include the director's gender.

```python
#!pip install plotly.express
import plotly.express as px 

q3_df = pd.DataFrame().assign(DirectorGender=final_df['DirectorGender'], bechdel_binary=final_df['bechdel_binary'], 
                              IsItGay=final_df['IsItGay'])
```

```python
fig = px.parallel_categories(q3_df, color='IsItGay')
fig.show()
```

As we can see from this parallel categories diagram using the director gender, Bechdel binary, and "is it gay" variables, LGBT films are much likely to pass the Bechdel test. 70% of LGBT films pass the Bechdel test compared to 55% of non-LGBT films. Also, 17% of films directed by women are LGBT films while 7% of films directed by men are LGBT films.
